### Data Cleaning

```python
#dropping unnecessary columns 
fifa_df = fifa_df.drop(columns = ['photoUrl','playerUrl','LongName'])
```

```python
#removing newline and star characters
fifa_df = fifa_df.replace('\n','', regex=True)
fifa_df = fifa_df.replace('★','', regex=True)
```

```python
#converting height to cm, weight to kg, hits contaning K, and columns to floats
def obj_to_float(val):
    if isinstance(val, str):
        if 'cm' in val:
            val = val.replace('cm', '')
            return float(val)
        elif "'" in val:
            val = val.split("\'")
            height = float(val[0])*(30.48) + float(val[1].replace('"', ''))*(2.54)
            return height 
        elif 'kg' in val:
            val = val.replace('kg', '')
            return float(val)
        elif 'lbs' in val:
            val = val.replace('lbs', '')
            weight = float(val)*0.45359237
            return weight
        elif 'K' in val:
            val = val.replace('K', '')
            hits = float(val)*1000
            return hits
        else:
            return float(val)
    else:
        return val
```

```python
fifa_df['Height']=fifa_df['Height'].apply(obj_to_float)
fifa_df['Weight']=fifa_df['Weight'].apply(obj_to_float)
fifa_df['Hits']=fifa_df['Hits'].apply(obj_to_float)
```

```python
def value_conversion(val):
    if isinstance(val, str):
        new_val = val.replace('€', '')
        if 'K' in new_val:
            new_val = float(new_val.replace('K', ''))*1000
        elif 'M' in new_val:
            new_val = float(new_val.replace('M', ''))*1000000
        return float(new_val)
```

```python
fifa_df['Wage']=fifa_df['Wage'].apply(value_conversion)
fifa_df['Value']=fifa_df['Value'].apply(value_conversion)
fifa_df['Release Clause']=fifa_df['Release Clause'].apply(value_conversion)
```

```python
#show columns that are of type object
fifa_df.select_dtypes(include='object')
```

```python
#converting object columns to integars
fifa_df[['W/F','SM', 'IR']] = fifa_df[['W/F','SM', 'IR']].astype(str).astype(int)
```

```python
#renaming columns
fifa_df = fifa_df.rename(columns={
    '↓OVA': 'OVA', 
    'Height': 'Height cm', 
    'Weight': 'Weight kg', 
    'W/F': 'Weak Foot',
    'SM': 'Skill Moves',
    'A/W': 'Attacker/Winger',
    'D/W': 'Defender/Winger',
    'IR': 'Intl Rep',
    'PAC': 'Pace/Speed',
    'SHO': 'Shooting',
    'PAS': 'Passing',
    'DRI': 'Dribbling',
    'DEF': 'Defense',
    'PHY': 'Physicality'})
```

```python
#separating joined column by month, day, and year
fifa_df[['Month Joined','Day Joined', 'Year Joined']] = fifa_df['Joined'].str.split(' ',expand=True)
fifa_df['Day Joined'] = fifa_df['Day Joined'].replace(',','', regex=True)
```

```python
#converting joined column into datetime
fifa_df['Joined'] = pd.to_datetime(fifa_df['Joined'], format='%b %d, %Y')
fifa_df['Joined'].head()
```
